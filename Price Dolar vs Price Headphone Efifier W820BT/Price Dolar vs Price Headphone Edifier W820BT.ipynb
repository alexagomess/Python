```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing data from excel
dados = pd.read_excel('C:/Users/alex-/Documents/GitHub/Python/Price Dolar vs Price Headphone Efifier W820BT/Dados_de_dolar_e_edifier.xlsx')
```

```python
dados.head()
```

```python
#declaration variables
mes = dados['Mês']
edifier = dados['Edifier']
dolar = dados['Dólar']
```

```python
#price edifier

plt.plot(mes,edifier)
plt.title("Preço do Headphone Edifier a partir de Maio")

plt.xlabel("Data")
plt.ylabel("Preço Edifier")

plt.show()
```

```python
#price dolar

plt.plot(mes,dolar)
plt.title("Preço do Dólar a partir de Maio")

plt.xlabel("Data")
plt.ylabel("Preço do Dólar")

plt.show()
```

```python
#price edifier vs price dolar

plt.scatter(edifier,dolar)
plt.title("Correlação entre o preço do headphone com o valor do dólar")

plt.xlabel("Valor do headphone")
plt.ylabel("Valor do dólar")

plt.show()
```

```python
# I don't see a strong correlaction the price dolar and growth price headphone edifier

eixox = dados['Mês']
eixoy_1 = dados['Edifier']
eixoy_2 = dados['Dólar']

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel("Data")
ax1.set_ylabel('Preço do Headphone', color=color)
ax1.plot(eixox, eixoy_1, color=color)


#instancia o segundo eixo com o mesmo eixo x
ax2 = ax1.twinx() 

color = 'tab:blue'
ax2.set_ylabel('Valor do Dólar', color=color)
ax2.plot(eixox, eixoy_2, color=color)


plt.show()
```
